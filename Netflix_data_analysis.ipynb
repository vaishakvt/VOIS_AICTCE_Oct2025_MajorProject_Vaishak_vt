```python

df = data.copy()

# Fill missing values
df['Director'].fillna('Unknown', inplace=True)
df['Cast'].fillna('Unknown', inplace=True)
df['Country'].fillna('Unknown', inplace=True)
df['Release_Date'].fillna('Unknown', inplace=True)
df['Rating'].fillna('Unknown', inplace=True)

# Convert 'Release_Date' to datetime format
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')

# Extract Year and Month from Release_Date
df['Year_Added'] = df['Release_Date'].dt.year
df['Month_Added'] = df['Release_Date'].dt.month

# Remove duplicates if any
df.drop_duplicates(inplace=True)

# Check cleaning summary
df.info()
```

```python
category_counts = df['Category'].value_counts()
```

```python
category_counts 
```

#### Movies vs TV Shows — Count Comparison

Goal: See how many Movies vs TV Shows Netflix has in total.

```python
category_counts.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Number of Movies vs TV Shows on Netflix')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

### See how Netflix’s content evolved each year

```python
trend = df.groupby(['Year_Added', 'Category']).size().unstack(fill_value=0)
```

```python
trend.plot(kind='line', figsize=(10,6))
plt.title('Netflix Content Trend Over the Years')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles Added')
plt.legend(title='Category')
plt.grid(True)
plt.show()
```

### Top 10 Countries by Content Production

Goal: See which countries produce the most Netflix content.

```python
# Handle multi-country entries by splitting and counting
country_split = df['Country'].dropna().apply(lambda x: [c.strip() for c in x.split(',')])
all_countries = pd.Series([country for sublist in country_split for country in sublist])

# Top 10 countries
top_countries = all_countries.value_counts().head(10)

# Plot
top_countries.plot(kind='bar', figsize=(10,6), color='lightgreen')
plt.title('Top 10 Content Producing Countries on Netflix')
plt.ylabel('Number of Titles')
plt.show()
```

### Top Genres on Netflix

Goal: Identify the most common genres

```python
# Split genres and flatten
genres_split = df['Type'].dropna().apply(lambda x: [g.strip() for g in x.split(',')])
all_genres = pd.Series([genre for sublist in genres_split for genre in sublist])

# Top 10 genres
top_genres = all_genres.value_counts().head(10)

# Plot
top_genres.plot(kind='bar', color='violet', figsize=(10,6))
plt.title('Top 10 Genres on Netflix')
plt.ylabel('Number of Titles')
plt.show()
```

From the analysis, International Movies appear as the most common genre on Netflix, followed by Dramas and Comedies. Among TV Shows, International TV Shows and Documentaries also have strong representation.
This indicates that Netflix focuses heavily on international and story-driven content, appealing to diverse global audiences.

##### Most Frequent Directors or Actors

```python
# Exclude 'Unknown' directors
top_directors = df[df['Director'] != 'Unknown']['Director'].value_counts().head(10)

# Plot
top_directors.plot(kind='bar', color='orange', figsize=(10,6))
plt.title('Top 10 Directors on Netflix')
plt.xlabel('Director')
plt.ylabel('Number of Titles')
plt.show()

```

#### Exploratory Data Analysis (EDA)

```python
# Count of Movies and TV Shows
category_counts = df['Category'].value_counts()

# Show the counts
print(category_counts)

# Visualize
category_counts.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Number of Movies vs TV Shows on Netflix')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

The analysis shows that Netflix has a higher number of Movies (5377) compared to TV Shows (2410).
This indicates that Netflix’s content library is still more movie-dominant, though TV Shows are steadily growing as part of its long-term engagement strategy.

### Trend Over the Years (Movies vs TV Shows)

```python
# Group by Year and Category
trend = df.groupby(['Year_Added', 'Category']).size().unstack(fill_value=0)

# Display a few years to check
print(trend.head())

# Line plot
trend.plot(kind='line', figsize=(10,6), marker='o')
plt.title('Netflix Content Trend Over the Years')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles Added')
plt.legend(title='Category')
plt.grid(True)
plt.show()
```

Netflix initially focused on movies, but as the platform matured, TV Shows became a major part of the content strategy, likely to retain subscribers longer with episodic content.

#### Top 10 Actors on Netflix

```python
# Split cast and flatten
cast_split = df['Cast'].dropna().apply(lambda x: [c.strip() for c in x.split(',')])
all_cast = pd.Series([actor for sublist in cast_split for actor in sublist])

# Remove 'Unknown' actors
all_cast = all_cast[all_cast != 'Unknown']

# Count top 10 actors
top_actors = all_cast.value_counts().head(10)

# Plot
top_actors.plot(kind='bar', color='lightcoral', figsize=(10,6))
plt.title('Top 10 Actors on Netflix')
plt.xlabel('Actor')
plt.ylabel('Number of Titles')
plt.show()
```

### Top Genres Trend Over the Years

```python
# Explode genres for trend analysis
df['Genres_List'] = df['Type'].dropna().apply(lambda x: [g.strip() for g in x.split(',')])
df_exploded = df.explode('Genres_List')

# Top 5 genres overall
top5_genres = df_exploded['Genres_List'].value_counts().head(5).index

# Group by Year and Genre
genre_trend = df_exploded.groupby(['Year_Added', 'Genres_List']).size().unstack(fill_value=0)

# Plot trend
genre_trend[top5_genres].plot(figsize=(10,6), marker='o')
plt.title('Top 5 Genres Trend Over the Years')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.legend(title='Genre')
plt.grid(True)
plt.show()
```

## WordCloud for Genres

```python
from wordcloud import WordCloud

text = " ".join(df_exploded['Genres_List'].dropna().astype(str))
wordcloud = WordCloud(width=1000, height=600, background_color='black', colormap='plasma').generate(text)

plt.figure(figsize=(10,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Popular Genres WordCloud on Netflix')
plt.show()
```
